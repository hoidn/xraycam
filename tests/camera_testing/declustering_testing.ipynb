```python
%matplotlib notebook
from xraycam.declustering import do_decluster
import numpy as np
import matplotlib.pyplot as plt
```

```python
size = 40
def test_array():
    testarr = np.random.rand(size, size)
    testarr[testarr < 0.9] = 0
    testarr -= 0.7
    testarr[testarr < 0] = 0
    testarr[testarr > 0] = 20
    testarr *= 100
    testarr = testarr.astype('uint8')
    testarr[:, size/2:] = 0
    return testarr
```

```python
def minimal_case():
    arr = np.zeros((5, 10), dtype = 'uint8')
    arr[2, 2] = 20
    arr[3, 2] = 20
    return arr
```

```python
def decluster_compare(genfunc = test_array, threshold = 5):
    fig, axes = plt.subplots(2, figsize = (12, 12))
    testarr = genfunc()
    plt.imshow(testarr, interpolation='none')
    dec = do_decluster(testarr, 5)

    for ax, arr in zip(axes, [testarr, dec]):
        im = ax.imshow(arr, interpolation = 'none')
    
```

```python
decluster_compare(minimal_case)
```
