```python
from xraycam.nbinit import *
```

```python
def plot_width_progression(dataruns, smooth = 5, rebin = 1):
    bragg = np.deg2rad(59.)
    energy = 2308.
    deltatheta = 5.2e-4/10
    deltaE = energy * deltatheta * np.cos(bragg)/np.sin(bragg)
    from xraycam.camalysis import fwhm
    lineouts = [dr.get_frame().get_lineout(smooth = 3, xmax = 1000)[1] for dr in dataruns]
    fwhmlist = list(map(lambda l: "FWHM: {:.4f} eV".format(deltaE * fwhm(l)), lineouts))

    salpha_width_study_lineouts = [dr.plot_lineout(peaknormalize = True, xmax = 1000, smooth = smooth,
                                        rebin = rebin, show = False, label = l)
                                   for dr, l in zip(dataruns, fwhmlist)]
    plt.show()
```
